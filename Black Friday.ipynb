```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
ohe_train= pd.get_dummies(newtrain)
```

```python
ohe_test = pd.get_dummies(newtest)
```

```python
sc.fit_transform()
```

```python
x = pd.DataFrame(sc.fit_transform(ohe_train.drop("Purchase", axis = 1)))
```

```python
y = train.Purchase
```

```python
scaled_test = pd.DataFrame(sc.fit_transform(ohe_test), columns=ohe_test.columns)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
```

```python
model = dtree.fit(x,y)
```

```python
predict = model.predict(scaled_test)
```

```python
sol = pd.DataFrame({"User_ID": test.User_ID, "Product_ID": test.Product_ID, "Purchase": predict})
```
